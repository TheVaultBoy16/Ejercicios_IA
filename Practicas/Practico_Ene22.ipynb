### Ejercicio 1 [0,75 ptos]:


Implementa las siguientes tres funciones de distancia que podrán ser utilizadas por el método kNN:

* **distancia_euclidea(v1,v2)**, cuya definición se presenta a continuación.

$$d_{E}(P,Q)={\sqrt {\sum _{i=1}^{n}(p_{i}-q_{i})^{2}}}$$


* **distancia_manhattan(v1,v2)**, cuya definición se presenta a continuación.

$$ d_{M}(P,Q)=\sum _{i=1}^{n}|p_{i}-q_{i}|$$

* **distancia_hamming(v1,v2)**, que se define como el número de atributos en los que difieren dos ejemplos dados.

**Nota:** Estas funciones recibirán dos vectores **v1** y **v2** de igual longitud que representan los valores de los atributos de dos ejemplos **sin sus valores de clasificación**.

```python
## Solución:

def distancia_euclidea(v1,v2):
    return math.sqrt((v2[0]-v1[0])**2 +(v2[1]-v1[1])**2)

def distancia_manhattan (v1,v2):
    return abs(v2[0]-v1[0]) + abs(v2[1]-v1[1])

def distancia_hamming(v1,v2):
    res=0
    for i,j in zip(v1,v2):
        if i != j:
            res+=1
    return res
```

### Ejercicio 2 [0,75 ptos]


Implementa una función **vecinos(distancia,dataset,ejemplo,k)** que dada una función de **distancia**, un **dataset**, un **ejemplo** y un valor entero **k** devuelva los **k** ejemplos en **dataset** más cercanos a **ejemplo** según la función **distancia**.

```python
## Solución:

def vecinos(distancia,dataset,ejemplo,k):
    for n in range(len(dataset)):
        dataset[n].append(distancia(dataset[n],ejemplo))
    return sorted(dataset, key=lambda e:e[3])[:k]


```

### Ejercicio 3 [0.5 ptos]:


A continuación, define una función **clasifica_kNN(distancia,dataset,ejemplo,k)** que recibiendo una función **distancia**, un **dataset**, un **ejemplo** y un valor entero **k** devuelva el valor de clasificación mayoritario en los **k** vecinos en **dataset** más cercanos a **ejemplo** según la función **distancia** recibida. Se espera que el valor de **k** a recibir sea impar para evitar empates.

```python
## Solución:

def clasifica_kNN(distancia,dataset,ejemplo,k):
    cd=dict()
    v=vecinos(distancia,dataset,ejemplo,k)
    for c in range(len(v)):
        clave=c[2]
        if clave not in cd:
            cd[clave]=0
        cd[clave]+=1
    return max(cd.items(),key=lambda e:e[1])
    
```

Ya que los conjuntos de datos con los que estamos trabajando son pequeños, realizaremos una evaluación del método utilizando validación cruzada. Implementa una función **kfolds(dataset,k)** que recibe un **dataset** y un valor entero **k** y divide el **dataset** en **k** subconjuntos, de tal manera que cada ejemplo del **dataset** debe pertenecer a sólo uno de los subconjuntos. La distribución de los ejemplos en los diferentes subconjuntos se realizará de manera aleatoria. Siempre que sea posible, los subconjuntos obtenidos deben tener la misma cardinalidad (esto no será posible cuando el número de ejemplos en el dataset no sea múltiplo de **k**).

```python
## Solución:

def kfolds(dataset,k):
    subconjuntos=[]
    if len(dataset) % k ==0:
        for i in range(0,len(dataset),k):
            subconjuntos.append(dataset[i: i+k])
    return subconjuntos

```

### Ejercicio 5  [0,75 ptos]:

Por último, define una función **cross_val_kNN(distancia,dataset,k1,k2)** que recibiendo una función **distancia**, un **dataset**, un valor entero **k1** que define el parámetro $k$ (número de vecinos a tener en cuenta) del algoritmo kNN y otro valor entero **k2** que define el número de folds en los que hay que dividir el conjunto de entrenamiento, realice lo siguiente:

1. Dividir el dataset en **k2** folds utilizando el método creado en el apartado anterior.
2. Para cada uno de los folds evaluar el rendimiento (la proporción de ejemplos que el método kNN clasifica correctamente) utilizando como datos de entrenamiento el resto de folds, el parámetro **k1** como número de vecinos a tener en cuenta y la función **distancia** como función de distancia para el método kNN.
3. Devolver la media de los rendimientos para los diferentes folds.

```python
## Solución:
def cross_val_kNN(distancia,dataset,k1,k2):
   rendimientos=[]
   d=0
   l=kfolds(dataset,k2)
   for f in l:
      for v in range(k1):
         d+=distancia(f,dataset[v])
      rendimientos.append(d/k1)
   return sum(r for r in rendimientos)/len(rendimientos)
```
