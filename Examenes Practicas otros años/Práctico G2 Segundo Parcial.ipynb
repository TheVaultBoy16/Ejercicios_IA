### APARTADO B (0.75 pto)

Se pide implementar una función `forward(e_x, ENTRADA, OCULTAS, SALIDA, PESOS, UMBRALES)` que, dado un ejemplo `e_x`, realice el paso de propagación hacia adelante, y devuelva `valores_in` y `valores_a`, correspondientes a dos diccionarios que almacenan, para cada neurona (dada por el índice), su valor de entrada y valor de activación, respectivamente.

Para ello:

```python
from copy import deepcopy

def forward(x, ENTRADA, OCULTAS, SALIDA, PESOS, UMBRALES):
    # 0. Guardamos en variables algunos elementos que pueden resultar relevantes
    entrada = deepcopy(ENTRADA)
    ocultas = deepcopy(OCULTAS)
    salida = deepcopy(SALIDA)
    pesos = deepcopy(PESOS)
    umbrales = deepcopy(UMBRALES)
    
    
    # 1.1 Inicializamos el diccionario que almacenará la salida de las neuonas. Para la capa de entrada corresponde a x
    
    valores_a = dict(zip(entrada, x))
    
    # 1.2 Inicializamos el diccionario que almacenará los valores de entrada de las neuronas. Inicialmente vacío, ya que no
    #     se necesita almacenar para la capa de entrada
    
    valores_in= dict()
    
    ### IMPLEMENTAR EL SIGUIENTE ALGORITMO ###
    
    # 2. Propagar las activaciones a través de las capas ocultas. 
    
    # 2.1 (Opcional) Almacenamos en una variable las neuronas de la última capa por la que se ha propagado, inicialmente la
    #                las neuronas de la capa de entrada:
    # ns_lprev = entrada[:] 
    
    
    # 2.2 Para cada capa oculta hacer:
    
        # 2.2.1 Para cada neurona de la capa, hacer
            
            # (-a-) in[i] ← umbral[i] + ∑ ( w[j,i] * a[j]); con j cada una de las neuronas de la capa anterior
            # (-b-) a[i] ← g_c(in[i])
        
        # 2.2.2 (Si se utiliza ns_lprev) Antes de pasar a la siguiente capa, actualizar la última capa visitada con las 
        #                                neuronas de la actual.
        # ns_lprev = ????
        
    
    # 3. Propagar las activaciones a través de la capa de salida
    
    # 3.1 Para cada neurona de la capa de salida , hacer:
    
        # in[i] ← umbral[i] + ∑ ( w[j,i] * a[j]); con j cada una de las neuronas de la capa anterior
        
        # a[i] ← g_c(in[i])
    
    # 4. Devolvemos los diccionarios con los valores de entrada y de salida
    
    return valores_in, valores_a
    
```
