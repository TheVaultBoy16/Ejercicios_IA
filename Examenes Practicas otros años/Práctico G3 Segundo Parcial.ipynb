Cuando un conjunto de entrenamiento tiene valores numéricos, no podemos aplicar directamente el algoritmo ID3 para construir un árbol de decisión. Para poder aplicar ese algoritmo, debemos transformar el conjunto de entrenamiento en otro con valores discretos. En los siguentes ejercicios vamos a implementar algunas funciones relacionadas con dicha transformación. Representaremos un conjunto de entrenamiento como un diccionario donde una de las claves tiene por valor 'Atributos' y representa los atributos y en el resto de los pares clave-valor, las claves son tuplas y el valor será 1 o 0. Nótese que con este sistema de representación con diccionarios, no puede haber ejemplos duplicados en un conjunto de entrenamiento. Ejemplos de conjuntos de entrenamiento:

```python
D1 = {('Cielo','Temperatura','Humedad','Viento'):'Atributos',
      ('Soleado', 17,  'NO',  12.1):0,
      ('Soleado', 26,  'NO',  12.5):0,          
      ('Nublado', 18,  'NO',  12.3):0,          
      ('Lluvia',  24,  'NO',  15.1):1,          
      ('Lluvia',  23,  'SI',  14.8):1,         
      ('Nublado', 27,  'SI',  12.2):0,          
      ('Soleado', 30,  'NO',  16.4):1,          
      ('Soleado', 22,  'SI',  13.7):1,          
      ('Lluvia',  21,  'SI',  16.9):1,          
      ('Soleado', 29,  'SI',  13.6):1,          
      ('Nublado', 20,  'NO',  17.3):1,          
      ('Nublado', 28,  'SI',  13.1):1,          
      ('Lluvia' , 19 , 'NO',  13.0):0}       

D2 = {(1.2, 'B1', 3):1,
      (3.1, 'B1', 7):0,          
      (1.4, 'B2', 4):1,          
      (1.5, 'B3', 5):1,
      ('AT1','AT2','AT3'):'Atributos',
      (1.4, 'B2', 2):1,          
      (2.0, 'B3', 6):0,          
      (2.3, 'B3', 10):0,          
      (3.6, 'B1', 13):0,          
      (4.0, 'B3', 1):1}    

D3 = {(1, 'T1'):0,
      (2, 'T2'):0,          
      (3, 'T3'):1,          
      (4, 'T4'):1,
      ('K1','K2'):'Atributos'}    
```

```python
def filtra_claves(dict,valor):
    d2={c: dict[c] for c in dict if valor in c}
    return {c: v for c,v in d2.items() if v==1}
```

```python
filtra_claves(D1,"Soleado")
```

### Ejercicio 1 [0.5 ptos]:
Escribir una función at_num(D) que recibe como entrada un conjunto de entrenamiento D y devuelve la lista de atributos que toman valores numéricos en D. __Nota:__ Los valores numéricos serán de tipo *float* o *int*, por tanto, para comprobar si un valor _x_ es de tipo numérico, puedes usar *isinstance(x,(int,float))*.

```python
# Solución

def at_num(D):
    res=[]
    for i in D:
        if not isinstance(i,(int,float)):
            res.append(i)
    return res


```

Puedes probar con los siguientes ejemplos

```python
at_num(D1)
# Salida:
# ['Temperatura', 'Viento']
```

```python
at_num(D2)
# Salida:
# ['AT1','AT3']
```

```python
at_num(D3)
# Salida:
# ['K1']
```

### Ejercicio 2 [1 pto]:
Dado un atributo, definimos un punto de corte como el valor medio de los valores _x1_ y _x2_ del atributo, siempre que _x1_ y _x2_ sean dos valores consecutivos (una vez ordenados) y la clasificación de los ejemplos correspondientes cambien. Se pide definir la función cortes(D,At) que reciba un conjunto de entrenamiento _D_ y un atributo de valores numéricos _At_ y devuelva los puntos de corte.

```python
# Solución:
def cortes(D,At):
    res=[]
    d2=dict()
    d2=sorted(D.items(),key=lambda e:e[1]) #ordenar por atributo
    for i in range(len(d2)-1):
        for c,v in d2.items():
            if d2[i][c] != d2[i-1][c]:
                res.append((d2[i][1]+d2[i-1][1])/2)
    return res
        
        
```

Ejemplos:

```python
cortes(D1,'Temperatura')
# Resultado:
# [19.5, 25.0, 27.5]
```

```python
cortes(D1,'Viento')
# Resultado:
# [13.05]
```

```python
cortes(D2,'AT1')
# Resultado:
# [1.75,3.8]
```

```python
cortes(D2,'AT3')
# Resultado:
# [5.5]
```

```python
cortes(D3,'K1')
# Resultado:
# [2.5]
```
