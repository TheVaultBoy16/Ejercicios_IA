### Ejercicio 1 (0,5 ptos):

Definir una función `accion_recomendada(s,pi)` que reciba como dato de entrada un estado *s* y una política *pi* y devuelva la acción recomendada para dicho estado *s* según la política *pi*.

Por ejemplo, si consideramos el estado $s1$ y la política *pi_1* la acción recomendada sería la acción $a1$. Si consideramos el estado $e1$ y la política *pi_3* la acción recomendada sería la acción $acc2$

```python
# Solución:
def accion_recomendada(s,pi):   # dado un s y una politica, buscar en esa politica la accion asociada a ese s
    ps=pi[0]
    pa=pi[1]
    return pa[ps.index(s)]
```

```python
accion_recomendada("s1",pi_1)
# Respuesta esperada: "a1"
```

```python
accion_recomendada("s4",pi_2)
# Respuesta esperada: "a2"
```

```python
accion_recomendada("e1",pi_3)
# Respuesta esperada: "acc_2"
```

```python
accion_recomendada("e1",pi_4)
# Respuesta esperada: "acc_1"
```

### Ejercicio 2 (0.5 ptos)
Define una función `ep(MDP,s,pi)` que tome como entrada un proceso de decisión de Markov, un estado $s$ y una política $pi$ y devuelva una lista de pares $(s',prob)$ donde los $s'$ son los estados a los que se puede llegar desde $s$ tras aplicar la acción que determina la política $pi$ y prob es la probabilidad de llegar a ese estado. Por ejemplo si tomamos el proceso de decisión MDP_1, el estado s2 y la política pi_2, vemos que la acción recomendada para s2 por esa política es a3 y que si aplicamos la acción a3 a s2 llegamos a
* s1 con probabilidad 0.9
* s4 con probabilidad 0.1

Por tanto la salida de la función debe ser [('s1', 0.9), ('s4', 0.1)]

```python
# Solución:
def ep(MDP,s,pi):
    a= accion_recomendada(s,pi)
    l=[]
    for i in MDP[1]:
        if i[0] == s and  i[1] == a:
            l=i[2: ]
    return l
```

```python
ep(MDP_1,'s4',pi_1)
# Respuesta esperada: [('s1', 0.3), ('s2', 0.1), ('s3', 0.1), ('s4', 0.7)]
```

```python
ep(MDP_2,'e1',pi_3)
# Respuesta esperada: [('e2', 0.3), ('e3', 0.7)]
```

```python
ep(MDP_2,'e2',pi_4)
# Respuesta esperada: [('e1', 1)]
```

### Ejercicio 3 (0.75 pto):
Definir una función `valoracion_secuencia(MDP,sec,desc)` que reciba como dato de entrada un proceso de decisión de Markov *MDP*, una secuencia de estados *sec* y un factor de descuento *desc* y devuelva la valoración de dicha secuencia de estados: 

$$V([q_0 q_1 q_2 ...])=R(q_0)+ \gamma R(q_1) + \gamma^2 R(q_2) + ...$$

Por ejemplo, si consideramos el problema $MDP\_1$, la secuencia $s_1, s_2, s_3, s_4$ y un factor de descuento 0.5 el resultado de la valoración sería:

$$V([s_1, s_2, s_3, s_4])=R(s_1)+ \gamma R(s_2) + \gamma^2 R(s_3) + \gamma^3 R(s_4)=-0.05 + 0.5(-0.05) + 0.5^2(-1)+ 0.5^3(1)=-0.2$$


```python
# Solución
def valoracion_secuencia(MDP,sec,desc):
    v = 0
    x = 0
    while x < len(sec):
        for y in MDP[0]:
            if sec[x] in y:
                v += y[1] * desc**x
        x+=1
    return v
```

```python
valoracion_secuencia(MDP_1,['s1','s2','s3','s4'],0.5)
# Respuesta esperada: -0.2
```

```python
valoracion_secuencia(MDP_1,['s1','s2','s1','s2'],0.25)
# Respuesta esperada: -0.06640625
```

```python
valoracion_secuencia(MDP_2,['e1','e2','e2'],0.5)
# Respuesta esperada: -0.375
```

```python
valoracion_secuencia(MDP_2,['e2','e2','e2'],0.9)
# Respuesta esperada: -0.271
```

### Ejercicio 4 (0.75 ptos):

Definir una función `probabilidad_secuencia(MDP,sec,pi)` que reciba como dato de entrada un proceso de decisión de Markov *MDP*, una secuencia de estados *sec* y una política *pi* y devuelva la probabilidad de dicha secuencia de estados bajo la política *pi*.

Por ejemplo, si consideramos el problema $MDP\_1$, la secuencia $[s1, s2, s3, s4]$ y la política *pi_1* (que asigna la acción $a1$ a todos los estados), la probabilidad resultante sería:

$$0.8 \times 0.1 \times 0.8=0.064$$

Ya que 0.8 es la probabilidad de pasar al estado $s2$ desde el estado $s1$ aplicando la acción recomendada por la política *pi_1* para $s1$, 0.1 es la probabilidad de pasar al estado $s3$ desde el estado $s2$ aplicando la acción recomendada por la política *pi_1* para $s2$ y 0.8 es la probabilidad de pasar al estado $s4$ desde el estado $s3$ aplicando la acción recomendada por la política *pi_1* para $s3$.

**Nota**: Se recomienda el uso de la función *ep(MDP,s,pi)* definida en el Ejercicio 1.


```python
# Solución:
def probabilidad_secuencia(MDP,sec,pi):
    p=1
    x=0
    while x < len(sec) - 1:
        estado_actual = sec[x]
        estado_siguiente = sec[x + 1]
        probabilidades = ep(MDP, sec[x], pi)
        prob_cero = any(estado_siguiente in estado for estado in probabilidades)
        if(prob_cero):
            for i in probabilidades:
                if (estado_siguiente in i):
                    p = p * i[-1]
        else:
            p = 0
        x += 1
    return p
```

### Ejercicio 5 (1 pto):

Definir una función `valoracion_respecto_politica(MDP,pi,descuento,k)` que aplica el método de iteración de valores *con una política fija* ($k$ iteraciones) para calcular la valoración $V^{\pi}$ (dicha valoración debe devolverse como un diccionario que a cada estado $s$ le asocia el valor $V^{\pi}(s)$ calculado). El algoritmo se describe en el punto 2 de la diapositiva 18 del tema 4. 

```python
# Solución:

def valoracion_respecto_politica(MDP,pi,descuento,k):
    parentesis=0
    Rdict=dict(MDP[0])   # consejo: crear la función R como un diccionario que 
                 # a cada estado le asigna su recompensa
        
    estados=[x[0] for x in MDP[0]]  # lista de estados
    
    V = {e:0 for e in estados}
    for _ in range(k):
        V1 = V.copy()
        for s in estados:
            probabilidades=ep(MDP,s,pi)
            for i in probabilidades:
                parentesis += (i[1]*V1[i[0]])
            V[s] = (Rdict[s]+descuento*(parentesis))
    return V
```
